```python
data = {
    'Department': ['HR', 'IT', 'Finance', 'IT', 'HR', 'Finance', 'IT'],
    'Salary': [40000, 50000, 60000, 55000, 42000, 65000, 52000],
    'Age': [25, 30, 35, 28, 26, 40, 32]
}
df = pd.DataFrame(data)
```

```python
df
```

groups "department" column if categorized

mean of "salary" is taken for all departments

```python
print(df.groupby("Department")["Salary"].mean())  # ✅ Group by 'Category' and sum 'Sales'
```

sum of "salary" and mean of "age" is taken for the grouped departments

```python
df.groupby('Department').agg({'Salary':'sum', 'Age':'mean'})
```

stores in csv back


```python
df.to_csv("output.csv", index=False)
```

```python
df1 = pd.DataFrame({'ID': [1,2,3], 'Name': ['Alice','Bob','Charlie']})
df2 = pd.DataFrame({'ID': [2,3,4], 'Score': [90, 85, 70]})
```

```python
df1
```

```python
df2
```

pd.merge(firsttable,second table,on=common key,how=type of join)

```python
pd.merge(df1,df2,on="ID",how="inner")
```

left keeps first table and whatever matches from second table

gives NaN for not matches


```python
pd.merge(df1,df2,on="ID",how="left")
```

keeps second table as it is and whatever matches from first(left)table

```python
pd.merge(df1,df2,on="ID",how="right")
```

keeps all rows of both tables

```python
pd.merge(df1,df2,on="ID",how="outer")
```
